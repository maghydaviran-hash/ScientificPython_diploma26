Write a function to solve a quadratic equation. Use the formula with the discriminant, only find the real solutions, if they don't exist print some message.

Hint: for the square root just use the power 0.5 to avoid having to call a special function which we haven't seen yet.

```python
def sol_quad(a, b, c):
    """ Find solutions for a* x**2 + b* x + c = 0 """
    if ( b**2- 4*a*c < 0 ):
        print('There are no real solutions')
    else :
        xp = (1 / (2*a) ) * ( -b + (b**2- 4*a*c)**0.5 )
        xn = (1 / (2*a) ) * ( -b - (b**2- 4*a*c)**0.5 )
    return xp, xn

print(sol_quad(1, 3, 1))
```

### **Exercise 5:**
Create a function that returns a list of all pairs of factors (as tuples) of a given number using list comprehension.

```python
def pairs_factors(n):
    l = []
    for i in range(1, n):
        if ( n%i == 0 and not((n/i, i) in l)):
            l.append((i, n/i))
    return l

print(pairs_factors(100))
```

### **Exercise 7:**
Write a function to check if a string (containing spaces and capital letters, but no punctuation marks) is a palindrome (use help to see which functions are available for a string). 


```python
def palindrome(word):
    processed_word = word.replace(" ", "").lower()

    splited = list(processed_word)
    is_a_palindrome = []
    #print(splited)
    
    for i in range(len(splited)):
        if splited[i] == splited[-(i + 1)]:
            is_a_palindrome.append(1)
        else:
            is_a_palindrome.append(0)
    #print(is_a_palindrome)
    
    if ([1]*len(splited) == is_a_palindrome):
        return print('The word is a palindrome')
    else:
        return print('The word is not a palindrome')

        
palindrome("amarama J, amarama")
```

Write a function to find a character with a maximum number of occurences in a string. Return a tuple with that character and the number. You can ignore ties. Only use string's inbuilt functions and the function `max` (use help to see how to use it).

```python
def f_character(str1):

    max_char = ''
    max_count = 0

    for char in set(str1):
        count = str1.count(char)

        if ( count > max_count ) :
            max_count = count
            max_char = char
    
    return (max_char, max_count)

f_character('sijsssssssogfjreopjua')
        
```

### **Exercise 9:**
Write a function that return the first N prime numbers using the Sieve of Eratosphenes algorithm

https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes

```python
def eratosphenes(n):
    l = [x for x in range(2, n)]
    for i in l :
        if ( i**2 > n ):
            break
        else:
            temporal = [x for x in range(i + i , n, i)]
            for j in temporal:
                try:
                    l.remove(j)
                except ValueError:
                    pass
    return l

print(eratosphenes(35))
```
